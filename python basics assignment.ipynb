```python
#2.What is an interpreter in Python?
#Ans:-
#In Python, an interpreter is a program that reads and executes Python code. Unlike compiled languages, where code is translated into machine code before execution, Python code is executed line-by-line by the interpreter.
#How the Python Interpreter Works
#When you run a Python script, the interpreter performs several steps:
#Lexical Analysis: The source code is broken down into tokens, which are the smallest units of meaningful code.
#Parsing: The tokens are analyzed to create an Abstract Syntax Tree (AST), which represents the hierarchical structure of the code.
#Compilation to Bytecode: The AST is compiled into bytecode, a lower-level representation of the code that is platform-independent.
#Execution by Python Virtual Machine (PVM): The bytecode is executed by the PVM, which translates it into machine code that the computer's hardware can execute.
#This process allows Python to be platform-independent, as the same bytecode can be executed on any system with a compatible interpreter
```

```python
#5.What is mutability in Python?
#Ans:-
#In Python, mutability refers to whether an object can be modified after it is created. Understanding the distinction between mutable and immutable objects is crucial for writing efficient and bug-free code.
#Mutable Objects
#Mutable objects can be changed after their creation. When you modify a mutable object, its identity (memory address) remains the same, but its content can be altered.
#Examples of mutable types:
#Lists: You can change, add, or remove elements.
#Dictionaries: You can add, modify, or delete key-value pairs.
#Sets: You can add or remove elements.

#Immutable Objects
#Immutable objects cannot be changed after their creation. Any operation that seems to modify an immutable object actually creates a new object.
#Examples of immutable types:
#Numbers: Integers, floats, and complex numbers
#Strings: You cannot change individual characters.
#Tuples: You cannot modify, add, or remove elements.
```

```python
#6.Why are lists mutable, but tuples are immutable?
#Ans:-
#Why Lists Are Mutable:-
#Lists are designed to be flexible and dynamic.
#Their mutability allows for operations like adding, removing, or changing elements, making them suitable for scenarios where the data collection needs to change over time.
#This flexibility comes at the cost of additional memory overhead and potentially slower performance due to the need to manage dynamic resizing and data integrity.

#Why Tuples Are Immutable:-
#Tuples are intended to represent fixed collections of items. Their immutability ensures that the data remains constant throughout the program, providing several benefits:
#Memory Efficiency: Tuples consume less memory than lists because they have a fixed size and do not require additional space for dynamic resizing. This makes them more efficient for storing large amounts of data.
#Performance: The immutability of tuples allows for faster access times compared to lists, as the system can optimize operations knowing that the data will not change.
#Hashability: Tuples can be used as keys in dictionaries or elements in sets because their contents cannot change, ensuring their hash value remains constant. In contrast, lists cannot be used in this way due to their mutability. 
#Safety: Immutability prevents accidental modifications, leading to more predictable and bug-free code, especially in concurrent programming environments.
```

```python
#14.Describe a scenario where a while loop is more suitable than a for loop.
#Ans:-
#Scenario: User Authentication with Unknown Number of Attempts
#Imagine you're developing a system that requires users to enter a password.
#You want to allow multiple attempts but stop once the correct password is entered. 
#Since the number of attempts is unknown and depends on user input, a while loop is appropriate:
correct_password = "secure123"
attempts = 0
max_attempts = 3
while attempts < max_attempts:
    user_input = input("Enter password: ")
    if user_input == correct_password:
        print("Access granted.")
        break
    else:
        print("Incorrect password. Try again.")
        attempts += 1

if attempts == max_attempts:
    print("Maximum attempts reached. Access denied.")

#In this example, the loop continues to prompt the user for the password until the correct one is entered or the maximum number of attempts is reached. 
#The number of iterations is not fixed in advance, making the while loop the suitable choice.

#Why Not a for Loop?
#A for loop is typically used when the number of iterations is known beforehand, such as iterating over a fixed range or a collection.
#In the user authentication scenario, since the number of attempts is not predetermined and depends on user input, a for loop would not be appropriate without introducing unnecessary complexity.
```

```python
#6.Write a function to demonstrate the behavior of mutable and immutable arguments.
#code>>
def demonstrate_mutability():
    # Immutable argument (integer)
    num = 10
    print(f"Original num: {num} (ID: {id(num)})")

    def modify_immutable(n):
        print(f"Before modification: ID = {id(n)}")
        n += 5
        print(f"After modification: ID = {id(n)}; n = {n}")
        return n

    num = modify_immutable(num)
    print(f"Outside function: num = {num} (ID: {id(num)})\n")

    # Mutable argument (list)
    lst = [1, 2, 3]
    print(f"Original list: {lst} (ID: {id(lst)})")

    def modify_mutable(lst):
        print(f"Before modification: ID = {id(lst)}")
        lst.append(4)
        print(f"After modification: ID = {id(lst)}; lst = {lst}")
        return lst

    lst = modify_mutable(lst)
    print(f"Outside function: lst = {lst} (ID: {id(lst)})")

demonstrate_mutability()
```

```python
#7. Write a program that performs basic arithmetic operations on two user-input numbers.
#code>>
# Function to perform basic arithmetic operations
def basic_arithmetic_operations():
    try:
        # Taking user input for two numbers
        num1 = float(input("Enter the first number: "))
        num2 = float(input("Enter the second number: "))

        # Performing and displaying the results of arithmetic operations
        print(f"\nAddition: {num1} + {num2} = {num1 + num2}")
        print(f"Subtraction: {num1} - {num2} = {num1 - num2}")
        print(f"Multiplication: {num1} * {num2} = {num1 * num2}")

        # Handling division by zero
        if num2 != 0:
            print(f"Division: {num1} / {num2} = {num1 / num2}")
        else:
            print("Division: Cannot divide by zero.")

    except ValueError:
        print("Invalid input. Please enter numeric values.")

# Calling the function to execute
basic_arithmetic_operations()
```

```python
#9. Write a Python program to convert user input from string to integer, float, and boolean types.
#code>>
def convert_input():
    user_input = input("Enter a value: ")

    # Convert to integer
    try:
        int_value = int(user_input)
        print(f"Converted to integer: {int_value}")
    except ValueError:
        print("Cannot convert to integer.")

    # Convert to float
    try:
        float_value = float(user_input)
        print(f"Converted to float: {float_value}")
    except ValueError:
        print("Cannot convert to float.")

    # Convert to boolean
    bool_value = bool(user_input)
    print(f"Converted to boolean: {bool_value}")

convert_input()
```

```python
#14.Write a program to reverse a string using a while loop.
#code>>
def reverse_string(input_string):
    reversed_string = ""
    index = len(input_string) - 1

    while index >= 0:
        reversed_string += input_string[index]
        index -= 1

    return reversed_string

# Example usage
original_string = "Hello, World!"
reversed_str = reverse_string(original_string)
print("Original String:", original_string)
print("Reversed String:", reversed_str)
```

```python
#15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.
#code>>
# Prompt the user to enter a number
num = int(input("Enter a number to find its factorial: "))

# Initialize variables
fact = 1
i = 1

# Check if the number is negative
if num < 0:
    print("Factorial is not defined for negative numbers.")
elif num == 0:
    print("The factorial of 0 is 1.")
else:
    # Calculate factorial using while loop
    while i <= num:
        fact *= i
        i += 1
    print(f"The factorial of {num} is: {fact}")
```
